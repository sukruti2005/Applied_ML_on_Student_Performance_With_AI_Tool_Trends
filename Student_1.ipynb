```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

inertias = []
sil_scores = []
db_scores = []
ch_scores = []

k_range = range(2, 12)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_train_scaled)

    inertias.append(km.inertia_)
    sil_scores.append(silhouette_score(X_train_scaled, labels))
    db_scores.append(davies_bouldin_score(X_train_scaled, labels))
    ch_scores.append(calinski_harabasz_score(X_train_scaled, labels))

fig, axes = plt.subplots(2, 2, figsize=(13, 8))


axes[0, 0].plot(k_range, inertias, 'o-', lw=2)
axes[0, 0].set_title('Elbow Method (Inertia)')
axes[0, 0].set_xlabel('Number of clusters (k)')
axes[0, 0].set_ylabel('Inertia')


axes[0, 1].plot(k_range, sil_scores, marker='o')
axes[0, 1].set_title('Silhouette Score')
axes[0, 1].set_xlabel('k')
axes[0, 1].set_ylabel('Score')


axes[1, 0].plot(k_range, db_scores, marker='o')
axes[1, 0].set_title('Davies-Bouldin Score')
axes[1, 0].set_xlabel('k')
axes[1, 0].set_ylabel('Score')


axes[1, 1].plot(k_range, ch_scores, marker='o')
axes[1, 1].set_title('Calinski-Harabasz Score')
axes[1, 1].set_xlabel('k')
axes[1, 1].set_ylabel('Score')

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd


df = pd.read_csv('ai_impact_student_performance_dataset.csv')


FEATURES = ['study_hours_per_day', 'ai_usage_time_minutes', 'attendance_percentage']
X = df[FEATURES]


X_train, _ = train_test_split(X, test_size=0.3, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)


dbscan = DBSCAN(eps=0.7, min_samples=5)
DB_labels = dbscan.fit_predict(X_train_scaled)


n_clusters_db = len(set(DB_labels)) - (1 if -1 in DB_labels else 0)
n_noise = (DB_labels == -1).sum()

print(f'DBSCAN Result:')
print(f'Estimated number of clusters: {n_clusters_db}')
print(f'Estimated number of noise points: {n_noise}')
print(f'Cluster sizes: {dict(zip(*np.unique(DB_labels, return_counts=True)))}')

)
if n_clusters_db > 1:
    non_noise = DB_labels != -1
    score = silhouette_score(X_train_scaled[non_noise], DB_labels[non_noise])
    print(f'Silhouette Score: {score:.4f}')


fig, ax = plt.subplots(figsize=(10, 6))

unique_labels = set(DB_labels)
colors = ['red', 'blue', 'green', 'yellow', 'purple']

for lbl, color in zip(unique_labels, colors):
    mask = DB_labels == lbl
    name = 'Noise' if lbl == -1 else f'Cluster {lbl}'

    ax.scatter(
        X_train_scaled[mask, 0],
        X_train_scaled[mask, 1],
        label=name
    )

ax.set_title('DBSCAN Clustering (Student Data)')
ax.set_xlabel(FEATURES[0])
ax.set_ylabel(FEATURES[1])
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pandas as pd

df = pd.read_csv('ai_impact_student_performance_dataset.csv')


FEATURES = ['study_hours_per_day', 'ai_usage_time_minutes', 'attendance_percentage']
TARGET = 'passed'

X = df[FEATURES]


label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(df[TARGET])


X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.30,
    random_state=42,
    stratify=y_encoded
)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


C_grid = [0.01, 0.1, 1, 10, 100]
gamma_grid = [0.01, 0.1, 1, 10]

results_grid = np.zeros((len(gamma_grid), len(C_grid)))


for i, g in enumerate(gamma_grid):
    for j, c in enumerate(C_grid):
        model = SVC(kernel='rbf', C=c, gamma=g, random_state=42)
        model.fit(X_train_scaled, y_train)
        results_grid[i, j] = accuracy_score(y_test, model.predict(X_test_scaled))


fig, ax = plt.subplots(figsize=(9, 5))

sns.heatmap(
    results_grid,
    annot=True,
    fmt='.3f',
    cmap='YlGn',
    xticklabels=C_grid,
    yticklabels=gamma_grid,
    linewidths=0.5,
    linecolor='white'
)

ax.set_xlabel('C (Regularization)', fontsize=11)
ax.set_ylabel('Gamma (Kernel Width)', fontsize=11)
ax.set_title('SVM (RBF Kernel) — Hyperparameter Tuning', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

z_arr = np.linspace(-8, 8, 300)

fig, ax = plt.subplots(figsize=(10, 6))


activations = {
    'Sigmoid σ(z) = 1/(1+e^-z)': (1/(1 + np.exp(-z_arr)), 'pink'),
    'Tanh tanh(z)': (np.tanh(z_arr), 'green'),
    'ReLU max(0, z)': (np.maximum(0, z_arr), 'black'),
    'Leaky ReLU': (np.where(z_arr > 0, z_arr, 0.1*z_arr), 'purple')
}

for name, (vals, color) in activations.items():
    ax.plot(z_arr, vals, color=color, label=name, lw=2.5)


ax.axhline(0, lw=0.8, alpha=0.5)
ax.axvline(0, lw=0.8, alpha=0.5)


ax.set_xlabel('z (pre-activation)', fontsize=11)
ax.set_ylabel('a(z) (activation output)', fontsize=11)
ax.set_title('Activation Functions\n(Non-linearity in Deep Learning)', fontweight='bold')


ax.set_ylim(-1.5, 1.5)

ax.annotate('ReLU: most common\nin hidden layers',
            xy=(2, 2), xytext=(0, -1.2),
            arrowprops=dict(arrowstyle='->'),
            fontsize=9)

ax.legend(fontsize=9)

plt.suptitle('Deep Learning Fundamentals — Artificial Neuron', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 2, figsize=(15, 6))
ax = axes[0]
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')


circle = plt.Circle((5, 5), 1.2, color='#4C72B0', alpha=0.85)
ax.add_patch(circle)
ax.text(5, 5, 'z', ha='center', fontsize=13, fontweight='bold', color='white')


inputs = [('x1', 2.0, 7.5, 'w1'),
          ('x2', 2.0, 5.0, 'w2'),
          ('x3', 2.0, 2.5, 'w3')]

for name, x, y, w in inputs:
    ax.annotate('', xy=(3.8, 5), xytext=(x + 0.4, y),
                arrowprops=dict(arrowstyle='->', lw=2))

    ax.text(x, y, name, ha='center', va='center',
            fontsize=13, fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='#aec6e8'))


    ax.text((x + 3.8) / 2, (y + 5) / 2, w, fontsize=10, color='red')


ax.text(4.2, 3.4, '+ b', fontsize=12, fontweight='bold')


ax.annotate('', xy=(8.5, 5), xytext=(6.2, 5),
            arrowprops=dict(arrowstyle='->', lw=2.5))
ax.text(9.1, 5, 'y', fontsize=14, fontweight='bold')


ax.text(5, 9.2, 'Single Artificial Neuron', ha='center',
        fontsize=13, fontweight='bold')
ax.text(5, 8.4,
        'z = w1x1 + w2x2 + w3x3 + b\n y = σ(z)',
        ha='center', fontsize=10)


ax2 = axes[1]
z_arr = np.linspace(-5, 5, 300)

activations = {
    'Sigmoid σ(z)': (1/(1 + np.exp(-z_arr)), '#d39e00'),
    'Tanh tanh(z)': (np.tanh(z_arr), '#2ca02c'),
    'ReLU max(0, z)': (np.maximum(z_arr, 0), '#1f77b4'),
    'Leaky ReLU': (np.where(z_arr > 0, z_arr, 0.01*z_arr), '#8172b2')
}

for name, (vals, color) in activations.items():
    ax2.plot(z_arr, vals, lw=2.5, label=name, color=color)

ax2.axhline(0, lw=0.8)
ax2.axvline(0, lw=0.8)

ax2.set_xlabel('z (pre-activation)')
ax2.set_ylabel('y (activation output)')
ax2.set_title('Activation Functions\n(Non-linearity in Deep Learning)', fontweight='bold')

ax2.set_ylim(-1.5, 1.5)
ax2.legend()


ax2.annotate('ReLU: most common\nin hidden layers',
             xy=(2, 2), xytext=(0, -1.2),
             arrowprops=dict(arrowstyle='->'),
             fontsize=9)

plt.suptitle('Deep Learning Fundamentals - Artificial Neuron',
             fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```
